```python
from src.parser import Parser
from src.lexer import Tokenizer
from src.models_utils import build_lukera_lexeme, build_lukera_grammar, parsing_table_pandas, compute_first, compute_follow


file_path = "exemplos/00_basico.lk"
lexeme = build_lukera_lexeme()
grammar = build_lukera_grammar()


luvas_lexer = Tokenizer(file_path, lexeme)
tokens = luvas_lexer.tokenize()


for token in tokens:
    print(f"<{token.get_type()}, {token.get_value()}>")
```

```python
luvas_parser = Parser(tokens, grammar)
luvas_parser.parse()

first_set = compute_first(grammar)
follow_set = compute_follow(grammar, first_set)
tabela_de_analise_preditiva = parsing_table_pandas(grammar, first_set, follow_set)

display(tabela_de_analise_preditiva)
```
